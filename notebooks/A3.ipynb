## 2. Client — send 4096 bytes, repeat 10 times

Each iteration: open a fresh socket, connect, send 4096 bytes, close. The server's `recv(1024)` will print how many bytes it actually got on the first call.

````python
import socket

PORT = 5001
PAYLOAD = b"x" * 4096

def socketClient(hostname, port):
  tcp_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
  tcp_socket.connect((hostname, port))
  print("Connected to socket: ", hostname, ":", port)
  return tcp_socket

for i in range(10):
  s = socketClient("localhost", PORT)
  s.sendall(PAYLOAD)
  s.close()

# Your code here:
# - loop 10 times
# - for each: create a TCP socket, connect to ('localhost', PORT), sendall(PAYLOAD), close
# - watch the server's terminal — it prints how many bytes recv(1024) returned each time
````

## 3. Record what the server saw

Copy the 10 byte-counts from the server's terminal into the list below.

````python
received = [
    1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024, 1024
]

print(f"min={min(received)}, max={max(received)}, distinct={len(set(received))}")
print(f"all 1024? {all(n == 1024 for n in received)}")
````
